###### 6. Display the first 5 rows of the DB

```python
#%%sql sqlite:///sf311
#SELECT * FROM sf311 limit 5;

pd.read_sql('SELECT * FROM sf311 LIMIT 5', con)
```

####7. Show 10 of the sidewalk cleanings in the Haight Ashbury neighborhood

```python
#%%sql sqlite:///sf311
#SELECT * from sf311 where neighborhood like '%Haight Ashbury%' and request_type ='Sidewalk_Cleaning' limit 10;

pd.read_sql('SELECT * from sf311 where neighborhood like "Haight Ashbury%" and request_type ="Sidewalk_Cleaning" limit 10', con)
    
```

####8. Which type of complaint was most common? 

```python
#%%sql sqlite:///sf311
#SELECT category, count(*) complaints from sf311 GROUP BY category ORDER BY complaints DESC;
#Street and Sidewalk Cleaning

#----
pd.read_sql('SELECT category, count(*) complaints from sf311 GROUP BY category ORDER BY complaints DESC', con)
```

####9. Are there any neighborhoods where sidewalk cleaning requests aren't the most common? 

```python
#%%sql sqlite:///sf311
#SELECT request_type, neighborhood, count(*) requests FROM sf311 GROUP BY request_type, neighborhood ORDER BY requests DESC limit 5;
#all the neighborhoods except Mission and South of Market

pd.read_sql('SELECT request_type, neighborhood, count(*) requests FROM sf311 GROUP BY request_type, neighborhood ORDER BY requests DESC limit 5', con)
```
